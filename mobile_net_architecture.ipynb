```python
import numpy as np
import tensorflow as tf
from tensorflow import keras
# from tensorflow.keras import layers,Dense
from tensorflow.keras.models import Sequential,Model
from tensorflow.keras.layers import Dense,Input, Dropout,DepthwiseConv2D, Activation,Layer, Flatten, Conv2D, MaxPooling2D,BatchNormalization,GlobalAveragePooling2D
from tensorflow.keras.datasets import mnist,cifar10
```

mobile net architecture

```python
def depth_wise_convlution(x,filter,strides):
  # depth wise convlution
  x = DepthwiseConv2D(kernel_size=3,strides=strides,padding='same')(x)
  x = BatchNormalization()(x)
  x = Activation('relu')(x)
  # pointwise convlution
  x = Conv2D(filters=filter,kernel_size=1,strides=1,padding='same')(x)
  x = BatchNormalization()(x)
  x = Activation('relu')(x)
  return x
```

```python
def mobile_net(input_shape,num_classes):
  inputs = keras.Input(shape=input_shape)
  # convlutional layer
  x = Conv2D(filters=32,kernel_size=3,strides=2,padding='same')(inputs)
  x = BatchNormalization()(x)
  x = Activation('relu')(x)
  # layer2
  x = depth_wise_convlution(x,64,1)
  x= depth_wise_convlution(x,128,2)
  x= depth_wise_convlution(x,128,1)
  x=depth_wise_convlution(x,256,2)
  x=depth_wise_convlution(x,256,1)
  x=depth_wise_convlution(x,512,2)
  for _ in range(5):
    x=depth_wise_convlution(x,512,1)
  x = depth_wise_convlution(x,1024,2)
  x = depth_wise_convlution(x,1024,1)
  x = GlobalAveragePooling2D()(x)
  output = Dense(num_classes,activation='softmax')(x)
  model = Model(inputs=inputs,outputs=output)
  return model
```

```python
model = mobile_net((32,32,3),10)
model.summary()
```

```python
model.compile(optimizer='adam',loss='sparse_categorical_crossentropy',metrics=['accuracy'])
```

```python
(x_train,y_train),(x_test,y_test) = cifar10.load_data()
```

```python
x_train = x_train/255
x_test = x_test/255
x_train = x_train.reshape(-1,32,32,3)
x_test = x_test.reshape(-1,32,32,3)
```

```python
history = model.fit(x_train,y_train,epochs=10,validation_data=(x_test,y_test))
```
